# **Fianance data**
*(From Wikipedia using Panda)*

```python
import pandas as pd

def load_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    df = html[0]
    return df
```

```python
df = load_data()
df
```

```python
sector_unique = df['GICS Sector'].unique() 
sector_unique
```

```python
len(sector_unique)
```

# **List of SP500 symbols**

```python
list(df.Symbol)
```

```python
data = yf.download(
        tickers = list(df.Symbol),
        period = "ytd",
        interval = "1d",
        group_by = 'ticker',
        auto_adjust = True,
        prepost = True,
        threads = True,
        proxy = None
    )
```

```python
data
```

```python
df2 = pd.DataFrame(data['WAT'].Close)
df2['Date'] = df2.index
df2
```

# **Plot Closing Price of Query Symbol**

```python
import matplotlib.pyplot as plt

plt.fill_between(df2.Date, df2.Close, color='skyblue', alpha=0.3)
plt.plot(df2.Date, df2.Close, color='skyblue', alpha=0.8)
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

```python
import matplotlib.pyplot as plt

def price_plot(symbol):
  plt.fill_between(df2.Date, df2.Close, color='skyblue', alpha=0.3)
  plt.plot(df2.Date, df2.Close, color='skyblue', alpha=0.8)
  plt.xticks(rotation=90)
  plt.title(symbol, fontweight='bold')
  plt.xlabel('Date', fontweight='bold')
  plt.ylabel('Closing Price', fontweight='bold')
  return plt.show()
```

```python
price_plot('GBPUSD')
```

```python
for i in list(df.Symbol)[:6]:
  price_plot(i)
```
